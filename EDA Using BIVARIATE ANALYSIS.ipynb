```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

## 1. Scatter Plot [Numerical-Numerical]

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],size=tips['size'])
```

## 2.Bar Plot [Numerical-Categorical]

```python
titanic=pd.read_csv('train.csv')
```

```python
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

## 3.Box Plot [Numerical-categorical]

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

## 4.Distplot [Numerical-Categorical]

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
```

```python
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=True)
```

## 5.Heat Map [Categorical-Categorical]

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind='bar')
```

## 6.Cluster Map [Categorical-Categorical]

```python
 sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```
